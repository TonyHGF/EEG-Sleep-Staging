```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from read_edf import read_edf_file

file_path = "../database/sleep-cassette/SC4001E0-PSG.edf"
labels, data, frequency = read_edf_file(file_path)

if labels and data and frequency:
    print("Signal Labels:", labels)
    print("Data Shape:", [len(channel) for channel in data])
    print("Sample Frequency:", frequency)
else:
    print("Failed to read .edf file.")
```

```python
# from read_edf import plot_signals


def debug(signal_labels, signals, sample_frequency):
    num_channels = len(signal_labels)
    fig, axes = plt.subplots(num_channels, 1, sharex=True, figsize=(10, 2 * num_channels))
    for i in range(num_channels):
        total_duration = len(signals[i]) / sample_frequency
        time_axis = np.linspace(0, total_duration, len(signals[i]))
        print(time_axis.shape, signals[i].shape)
        axes[i].plot(time_axis, signals[i], label=signal_labels[i])
        axes[i].set_ylabel("Amplitude")
        axes[i].legend()
    axes[-1].set_xlabel("Time (seconds)")
    plt.tight_layout()
    plt.show()

debug(labels, data, frequency)
# plot_signals(labels, data, frequency)

# def plot_edf_signalaaa(signal_labels, signals, sample_frequency):
#     """
#     Plots the signals from an EDF file.

#     Parameters:
#     signal_labels (list): A list of signal labels.
#     signals (list): A list of signal arrays.
#     sample_frequency (int): The sample frequency of the signals.
#     """
#     num_signals = len(signals)
#     time_vector = np.arange(signals[0].size) / sample_frequency

#     # 设置图表大小
#     plt.figure(figsize=(12, 2 * num_signals))
# #     print("aaa\n")
#     # 遍历所有信号并绘制
#     for i, signal in enumerate(signals):
#         plt.subplot(num_signals, 1, i + 1)
#         plt.plot(time_vector, signal)
#         plt.title(signal_labels[i])
#         plt.xlabel('Time (seconds)')
#         plt.ylabel('Amplitude')
# #         break
#     plt.tight_layout()
#     plt.show()
    
    
# print("here")
# plot_edf_signals(labels, data, frequency)
```
